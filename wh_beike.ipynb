```python
import pandas as pd
import re
df = pd.read_csv("wh_beike2.csv",encoding = 'ANSI')
df.head(10)
```

```python
df.info()
```

```python
s='业主直租·福星惠誉国际城3期·2居室'
result = re.split(r'(.*?)·(.*?)·',s)[2]
print(result)
```

```python
df['小区'] = df['小区名称'].apply(lambda x : re.split(r'(.*?)·(.*?)·',x)[2])
```

```python
df.sample(10)
```

```python
df['房间大小'] = df['面积'].apply(lambda x : re.split(r'(.*?)㎡',x)[1])
```

```python
df.sample(10)
```

```python
df.info()
```

```python
df['附近地铁站'] = df['距地铁步行距离'].apply(lambda x : re.split(r'小区距(.*?)号线(.*?)约(.*?)米',x)[1])
df['距地铁距离'] = df['距地铁步行距离'].apply(lambda x : re.split(r'小区距(.*?)号线(.*?)约(.*?)米',x)[3])
```

```python
df.info()
```

```python
df.sample(10)
```

```python
data = df[['小区名称', '租金',  '行政区', '类型', '小区',  '房间大小', '面积', '附近地铁站', '距地铁距离']]
```

```python
data.sample(10)
```

```python
data['租金'].astype('float')
```

```python
data['大小'] = data['面积'].apply(lambda x : re.split(r'[\D]?(.*?)㎡',x)[1])
```

```python
data.sample(10)
```

```python
df1 = data[['小区名称', '租金',  '行政区', '类型', '小区',  '大小',  '附近地铁站', '距地铁距离']]
```

```python
df1.info()
```

```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
from pyecharts.charts import *
```

```python
df1['序号'] =range(1, df1.shape[0]+1)
```

```python
df1.sample()
```

```python
wuhan = df1.groupby('行政区',as_index=False)['序号'].count().sort_values('序号',ascending=False).reset_index(drop=True)
wuhan.loc[~wuhan['行政区'].str.endswith('区'),'行政区']=wuhan['行政区']+'区'
```

```python
df1['地区'] = df1['行政区'].apply(lambda x : re.split(r'(.*?)区',x)[1])
```

```python
df1.sample()
```

```python
wuhan = df1.groupby('地区',as_index=False)['序号'].count().sort_values('序号',ascending=False).reset_index(drop=True)
wuhan.loc[~wuhan['地区'].str.endswith('区'),'地区']=wuhan['地区']+'区'
```

```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
from pyecharts.charts import *
```

```python
# 绘制地图（区）
c = (
    Map(init_opts=opts.InitOpts(theme='dark', width='800px'))
    .add("房源数", [list(z) for z in zip(wuhan['地区'].to_list(), wuhan['序号'].to_list())], "武汉", label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="自如武汉房源数分布",
                                 subtitle='数据采集日期：2023年6月'), 
        visualmap_opts=opts.VisualMapOpts(max_=5000)
    )    
)
c.render('map2_wuhan.html')
```

```python
# 绘制柱状图
bar = (Bar(init_opts=opts.InitOpts(theme='dark', width='800px'))
            .add_xaxis(wuhan['地区'].to_list())
            .add_yaxis("", wuhan['序号'].to_list())
            .set_series_opts(label_opts=opts.LabelOpts(is_show=True, 
                                                       position='right',
                                                       font_style='italic'),
                            itemstyle_opts=opts.ItemStyleOpts(
                                color=JsCode("""new echarts.graphic.LinearGradient(1, 0, 0, 0, 
                                             [{
                                                 offset: 0,
                                                 color: 'rgb(255,99,71)'
                                             }, {
                                                 offset: 1,
                                                 color: 'rgb(32,178,170)'
                                             }])"""))
                            )
            .set_global_opts(
                title_opts=opts.TitleOpts(title="自如武汉房源数量",
                                         subtitle='数据采集日期：2023年6月'),
                xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=45)),
                legend_opts=opts.LegendOpts(is_show=True))
            .reversal_axis()
        )
bar.render('bar2_wuhan.html')
```

```python
# 租金箱线图
x_data = list(df1['地区'].unique())
y_data = [data[(df1['地区']==x)]['租金'].to_list() for x in x_data]

Box = (Boxplot(init_opts=opts.InitOpts(theme='dark', width='800px'))
       .set_global_opts(legend_opts=opts.LegendOpts(is_show=True),
                     xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=45)),
                     title_opts=opts.TitleOpts(title="自如武汉各区【房租】箱线图",
                                         subtitle='数据采集日期：2023年6月'),
                    )
      ) 
Box.add_xaxis(x_data)
Box.add_yaxis("", Box.prepare_data(y_data))
Box.render('boxplot_wuhan.html')
```
